```python
import pandas as pd
import os
```

```python
data_path = 'D:/Orange/Episafe/final-data'
```

```python
dataframes=[]
for i in  os.listdir(data_path):
    path = os.path.join(data_path,i).replace(os.path.sep,'/')
    dataframe = pd.read_csv(path)
    dataframes.append(dataframe)

final_data = pd.concat(dataframes,ignore_index=True)
```
