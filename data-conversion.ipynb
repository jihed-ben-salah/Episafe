```python
bucket_name = 'episafe-data'
prefix = 'train/1869/'
```

```python
# List objects in the bucket with the specified prefix
objects = s3_client.list_objects(Bucket=bucket_name, Prefix=prefix)

# Get the list of object keys
object_keys = [obj['Key'] for obj in objects['Contents']]

# Read Parquet files and create a list of dataframes
dataframes = []
for object_key in object_keys:
    s3_path = f's3://{bucket_name}/{object_key}'
    dataset = pq.ParquetDataset(s3_path)
    table = dataset.read()
    dataframe = table.to_pandas()
    dataframe['path'] = s3_path
    print(s3_path)
    dataframes.append(dataframe)
    

# Concatenate the list of dataframes into a single dataframe
combined_dataframe = pd.concat(dataframes, ignore_index=True)
```
