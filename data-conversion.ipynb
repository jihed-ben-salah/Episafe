```python
import os
import pyarrow.parquet as pq
```

```python
dat_path = "D:/Orange/Episafe_local/data/parquet/train/1869"
```

```python
parquet_files = [os.path.join(dat_path, file) for file in os.listdir(dat_path) if file.endswith(".parquet")]
```

```python
all_parquet =[]
for i in os.listdir(dat_path):
    path = os.path.join(dat_path, i).replace(os.path.sep,'/')
    parquet_files = [os.path.join(path, file).replace(os.path.sep,'/') for file in os.listdir(path) if file.endswith(".parquet")]
    all_parquet.append(parquet_files)
```

```python
flat_list = [item for sublist in all_parquet for item in sublist]
```

```python

done = ['000','001','002','003','004','005','006','007','008','009','010','012','013','014','015','016','017']
for folder in range(len(all_parquet)):
    for path_parquet in all_parquet[folder]:
        if path_parquet.split('/')[7] not in done:
            dataset = pq.ParquetDataset(path_parquet)
            table = dataset.read()
            dataframe = table.to_pandas()
            dataframe['path'] = path_parquet
            print(path_parquet)
            dataframe.to_csv(path_parquet.split('/')[6]+'_'+path_parquet.split('/')[7]+'.csv')
        else:
            print(path_parquet.split('/')[7]+' :done')
    done.append(path_parquet.split('/')[7])

# Concatenate the list of dataframes into a single dataframe
#combined_dataframe = pd.concat(dataframes, ignore_index=True)
```
